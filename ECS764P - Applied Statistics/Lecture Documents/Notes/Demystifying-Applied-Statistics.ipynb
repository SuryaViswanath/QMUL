#### Let's get a sense of these theorem's and how they help us in real world.


**Probability Theorem in Machine Learning**

In machine learning, understanding probability helps us make predictions and decisions under uncertainty. Lets dive into the key concepts with their mathematical formulations!

##### 1. **Addition Rule (for mutually exclusive events)**

If two events, \( A \) and \( B \), are mutually exclusive (which means they cannot occur at the same time), the probability of either of the event happening is the sum of their individual probabilities:

$$
P(A \ \text{or} \ B) = P(A) + P(B)
$$

**Example:** The probability of rolling a 2 or a 5 on a die is:

$$
P(2 \ \text{or} \ 5) = P(2) + P(5) = \frac{1}{6} + \frac{1}{6} = \frac{1}{3}
$$

##### 2. **Multiplication Rule (for independent events)**

When two events, \( A \) and \( B \), are independent (which means the occurrence of one does not affect the occurrence of the other), the probability of both events happening is the product of their probabilities:

$$
P(A \ \text{and} \ B) = P(A) \times P(B)
$$

**Example:** The probability of flipping a coin and getting heads twice in a row is:

$$
P(H \ \text{and} \ H) = \frac{1}{2} \times \frac{1}{2} = \frac{1}{4}
$$

##### 3. **Bayes' Theorem**

Bayes' Theorem allows us to update the probability of an event based on new evidence. Its particularly useful in machine learning for classification and decision-making problems:

$$
P(A|B) = \frac{P(B|A) \times P(A)}{P(B)}
$$

Where:
- \( P(A|B) \) is the probability of event \( A \) given that event \( B \) has occurred (posterior probability).
- \( P(B|A) \) is the probability of event \( B \) given \( A \) (likelihood).
- \( P(A) \) is the prior probability of \( A \).
- \( P(B) \) is the probability of event \( B \).

**Example:** In a medical diagnosis, if a test result is positive, Bayes Theorem helps calculate the actual probability of having the disease, given the accuracy of the test and the prevalence of the disease.

##### **How These Theorems Save Us in Machine Learning**

- **Addition Rule**: Helps in scenarios where we are assessing the probability of one of many possible outcomes, like classifying an input into one of several categories.
  
- **Multiplication Rule**: Critical for **Naive Bayes** classifiers, where we assume features are independent and calculate joint probabilities to make predictions.

- **Bayes' Theorem**: The backbone of **Bayesian Inference** used in models to update predictions as new evidence becomes available.

Probability isnt just a numberits a **lifesaver** when handling uncertainty in machine learning!

Preliminary concepts:

1. **Sets**: A set is a collection of numbers usually represented by logical condition (for instance $$\{ x\in \mathbb{N} \ | 5 < x \}$$ <br> The above equation reads: All natural numbers $x$ such that $5<x$
2. **Operations on Sets**

    In set theory, there are several fundamental operations that can be performed on sets. Here are the most common ones:
    
    1. **Union**
    
    The **union** of two sets \( A \) and \( B \) is the set of elements that are in either \( A \), \( B \), or both. It is denoted by $( A \cup B )$.
    
    $$ 
    A \cup B = \{ x \mid x \in A \text{ or } x \in B \} 
    $$
    
    *Example*:  
    Let \( A = \{1, 2, 3\} \) and \( B = \{3, 4, 5\} \).  
    Then, the union of \( A \) and \( B \) is:  
    $$ 
    A \cup B = \{1, 2, 3, 4, 5\} 
    $$
    
    2. **Intersection**
    
    The **intersection** of two sets \( A \) and \( B \) is the set of elements that are in both \( A \) and \( B \). It is denoted by $( A \cap B )$.
    
    $$ 
    A \cap B = \{ x \mid x \in A \text{ and } x \in B \} 
    $$
    
    *Example*:  
    Continuing with the sets \( A = \{1, 2, 3\} \) and \( B = \{3, 4, 5\} \),  
    the intersection of \( A \) and \( B \) is:  
    $$ 
    A \cap B = \{3\} 
    $$
    
    3. **Relative Complement** 
    
    The **relative complement** (or difference) of set \( B \) in set \( A \) is the set of elements that are in \( A \) but not in \( B \). It is denoted by \( A - B \) or $( A \setminus B)$.
    
    $$ 
    A - B = \{ x \mid x \in A \text{ and } x \notin B \} 
    $$
    
    *Example*:
    Using the same sets \( A = \{1, 2, 3\} \) and \( B = \{3, 4, 5\} \),  
    the relative complement of \( B \) in \( A \) is:  
    $$ 
    A - B = \{1, 2\} 
    $$
    
    4. **Cartesian Product**
    
    The **Cartesian product** of two sets \( A \) and \( B \) is the set of all ordered pairs \( (a, b) \) where $( a \in A )$ and $( b \in B )$. It is denoted by $( A \times B )$.
    
    $$ 
    A \times B = \{ (a, b) \mid a \in A \text{ and } b \in B \} 
    $$
    
    *Example*:  
    For sets \( A = \{1, 2\} \) and \( B = \{x, y\} \),  
    the Cartesian product \( A \times B \) is:  
    $$ 
    A \times B = \{ (1, x), (1, y), (2, x), (2, y) \} 
    $$

Let's take a fun example: Imagine we're playing a game with a die. All the possible outcomes are under one umbrella called the sample space, and the chance of getting any number is what we call the probability. Combining these two concepts is what we refer to as a **measure**. Think of it as measuring the likelihood of you rolling a six while your friend rolls a one and brags about it!

A measure usually satisfies three principles:

1. The probability of the sample space totals to 1, i.e., $( P(S) = 1 )$.
  
2. The probability of any event in the sample space can never be negative, i.e., $( P(E) \geq 0 )$. (Were not in a negative mood here!)

3. The additivity rule states that for any two mutually exclusive events (events that cannot happen at the same time), the probability of either event happening is the sum of their individual probabilities. Mathematically,
