# Computer Vision

Computer vision is the ability to teach the computers abilities to perceive information from an image. If your 2 year old can count the number of people and name them from your family group picture, so should the computers also. 

However, `Why is vision so difficult?`

Unlike human brains who have the ability to perceive the world in 3D, vision doesn't have the capabilities to perceive the information in 3D, it needs to rely on physics-based and probabilistic models to disambiguate between potential solutions.

Hence, the vision becomes an inverse problem where it relies on uncovering information from the source where there is insuffiecient information to fully understand the information and then answer.

Recreating the world as we see it using the images takes a lot of understanding, it is best put here as follows - _" how objects move and animate, how light reflects off their surfaces, is scattered by the at- mosphere, refracted through camera lenses (or human eyes), and finally projected onto a flat (or curved) image plane"_
