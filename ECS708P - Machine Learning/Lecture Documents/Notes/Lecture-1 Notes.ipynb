#### Machine learning can be classified into three categories as follows:

1. Supervised Learning
> System is trained on labeled data, meaning each training example is paired with the correct output (i.e., ground truth). The goal is for the system to generalize and make predictions on new, and unseen data.

2. Unsupervised Learning
> data has no labels or ground truth. The goal for the system is to discover hidden patterns, groupings, or structures from the data without any supervision.

3. Reinforcement Learning
> Trial and error, hopefully the first time you touch the hot pan, you know you shouldn't touch it
